```python
df1 = df.drop('PassengerId',axis ='columns')
df2 = df1.drop('SibSp',axis = 'columns')
df3 = df2.drop('Parch',axis='columns')
df4 = df3.drop('Cabin',axis ='columns')
df5 = df4.drop('Name_Initial', axis='columns')
df6 = df5.drop('Embarked', axis = 'columns')
df7 = df6.drop('Name', axis='columns')

df7 = df7.replace({
    'male': '0',
    'female':'1'
})
df7['Sex'] = df7['Sex'].astype(int)
df7.info()
df8 = df7.drop('Ticket',axis = 'columns')
df8
df9 = df8.drop('Age',axis='columns')
```

```python
X = df9.drop('Survived',axis='columns')
y = df9.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=23)
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
