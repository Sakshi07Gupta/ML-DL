```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
# Logistic Regression
lr = LogisticRegression(solver='liblinear',multi_class='ovr')
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
sc1 = cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), iris.data, iris.target,cv=3)
np.average(sc1)
```

```python
sc2 = cross_val_score(SVC(gamma='auto'), iris.data, iris.target,cv=3)
np.average(sc2)
```

```python
sc3 = cross_val_score(RandomForestClassifier(n_estimators=10), iris.data, iris.target,cv=3)
np.average(sc3)
```

```python
sc4 = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target, cv= 3)
np.average(sc4)
```
