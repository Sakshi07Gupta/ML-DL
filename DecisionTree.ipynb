```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
inputs
```

```python
target = df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder # sklearn libaray hain # preprocessing module hain
                                               # labelEncoder ek class hain  
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
# Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?
```

```python
model.predict([[2,1,0]])
```

```python
# Is salary of Google, Computer Engineer, Masters degree > 100 k ?
```

```python
model.predict([[2,1,1]])
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
inputs
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex	 = LabelEncoder()
le_Age	 = LabelEncoder()
le_Fare= LabelEncoder()
```

```python
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
inputs
```

```python
inputs_n = inputs.drop(['Pclass','Sex','Age','Fare'],axis='columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```
