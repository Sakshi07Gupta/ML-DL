```python
import math
```

```python
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(100)
```

```python
tanh(1)
```

```python
tanh(-56)
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(-2)
```

```python
relu(100)
```

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-19)
```

```python
leaky_relu(23)
```
