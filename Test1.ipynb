```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("insurance.csv")
df.head()
```

```python
df1 = pd.read_csv("test.csv")
df1.head()
```

```python
%matplotlib inline 
plt.xlabel('x-sample')
plt.ylabel('y-sample')
plt.scatter(df1.x,df1.y,color='purple',marker='.')
```

```python
new_y = df1.y
new_y
```

```python
model = linear_model.LinearRegression()
model.fit(df11,new_y)
```

```python
model.coef_
```

```python
model.intercept_
```
